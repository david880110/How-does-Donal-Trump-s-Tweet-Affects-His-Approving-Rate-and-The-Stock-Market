```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df=pd.read_csv("TrumpTweets.csv")
df.head()
```

```python
firstTweet = True
compound_list =[]
sentiments =[]
count =0

for index, row in df.iterrows():
    #Run Vader Analysis on each tweet
    sentence =row["text"]
    #Tweet datetime object
    new_date = row["created_at"]
    if(firstTweet == False):
        if(old_date != new_date):
            compound= np.mean(compound_list)
            compound_list =[]
            #Store the Sentiments
            sentiments.append({"Compound":compound,
                "Date" : old_date,
                "Count" : count})
            #Resetting count
            count =0
    else:
        firstTweet = False
    compound_list.append(analyzer.polarity_scores(sentence)["compound"])
    count +=1
    old_date = new_date
```

```python
#Put all data into dataframe and export to csv file
sentiments_df = pd.DataFrame.from_dict(sentiments)
sentiments_df.head()
```

```python
for index, row in sentiments_df.iterrows():
    #row['startdate']=row['startdate'].date()
    d= datetime.strptime(row['Date'], '%m/%d/%y')
    sentiments_df.loc[index, 'Date'] = d.strftime('%Y-%m-%d')
sentiments_df=sentiments_df.loc[(sentiments_df['Date'] > '2017-01-19') & (sentiments_df['Date'] < '2018-01-01')]
sentiments_df.tail()
```

```python
for index, row in sentiments_df.iterrows():
    d = datetime.strptime(str(row['Date']), '%Y-%m-%d')
    sentiments_df.loc[index, 'Date']= d.strftime('%m/%d/%Y')
sentiments_df.head()
```

```python
df1=pd.read_csv("approval_polllist.csv")
df1.head()
```

```python
df2=df1[['subgroup','startdate','pollster','approve']]
df2=df2.loc[(df2['pollster'] == 'Ipsos') & (df2['subgroup'] == 'All polls')]
df2.head()
```
